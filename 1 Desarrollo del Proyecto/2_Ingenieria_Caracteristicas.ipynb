```python
from pandasql import sqldf
import pandas as pd
import numpy as np
import unidecode
```

```python
# directorios (path):
path = "gdrive/MyDrive/my_notebooks/datasets/"
```

```python
df = pd.read_excel(path + 'Banafrut.xlsx', sheet_name='Base')
df
```

```python
columns = df.columns
columns =[unidecode.unidecode(c).replace(' ','_').replace('(','').replace(')','') for c in columns]
df.columns = columns
df
```

```python
datatypes = {'Ano': 'int64',
             'Semana': 'int64',
             'Zona': 'category',
             'Finca': 'category',
             'municipio': 'category',
             'temp_media': 'float64',
             'Precipitacion': 'float64',
             'Humedad_relativa': 'float64',
             'Velocidad_del_viento': 'float64',
             'Peso_total_del_racimo_kg': 'float64'}

for col, dtype in datatypes.items():
    df[col] = df[col].astype(dtype)

df.info()
```

Se crea variable Ano_Semana como identificador

```python
df['Ano_Semana']=df['Ano']*100+df['Semana']
df['Ano_Semana']
```

```python
df_aux = df['Ano_Semana'].value_counts().reset_index()['Ano_Semana']
df_aux = pd.DataFrame(sorted(df_aux.to_list()),columns=['Ano_Semana'])
df_aux
```

```python
# Asignar a cada Ano_Semana un id unico para que sirva como secuencia
df_aux = df_aux.reset_index()
df_aux['index'] = df_aux['index']+1
df_aux
```

Se crea variable Semana_seq para cada finca de forma que se tenga un id para cada Semana

```python
df['Semana_seq'] = df['Ano_Semana'].apply(lambda x: df_aux[df_aux['Ano_Semana']==x]['index'].tolist()[0])
df
```

Ahora procedemos a aplicar el rezago en las variables:
- temp_media
- Precipitacion
- Humedad_relativa
- Velocidad_del_viento

```python
# Definir funcion generadora de rezagos con SQL utilizando 'Semana_seq' como variable de ordenamiento
def generarSQLRezago(colname,n_resagos):
    sql = ''
    for i in range(1, n_resagos+1):
        semana = n_resagos-i
        sql = sql+f'''lag({colname},{i},null) over (partition by zona,finca,municipio order by Semana_seq asc) {colname}_lag{i},\n'''
    return sql[0:len(sql)-2]
```

```python
n_rezagos = 51
df_rezagos = sqldf(f'''select ano,
                    semana,
                    zona,
                    finca,
                    municipio,
                    semana_seq,
                    ano_semana,
                    {generarSQLRezago('temp_media',n_rezagos)},
                    {generarSQLRezago('Precipitacion',n_rezagos)},
                    {generarSQLRezago('Humedad_relativa',n_rezagos)},
                    {generarSQLRezago('Velocidad_del_viento',n_rezagos)},
                    Peso_total_del_racimo_kg
from df ''')

df_rezagos.tail()
```
